V1 [Januar – September, Oktober (Split A)] Machen Sie sich Sorgen darüber, dass so viele Flüchtlinge nach Deutschland kommen, oder machenSie sich keine Sorgen? = 독일로 많은 난민들이 오는 것에 대하여 걱정하시나요? 아니면 걱정하지 않으시나요?

V1b [Oktober (Split B), November, Dezember]Machen Sie sich Sorgen darüber, dass so viele Flüchtlinge in Deutschland sind, oder machen Sie sichkeine Sorgen? = 독일에 이렇게 많은 난민들이 있는 것에 대하여 걱정하시나요? 아니면 걱정하지 않으시나요

V1과 V1b는 같은 의미의 질문으로, 질문 시기의 차이로 인해 표현만 다를 뿐이므로, 하나의 통합된 열로 결합해야한다. 

```python
value_counts_v1=data['V1'].value_counts()
print(value_counts_v1)
value_counts_v1b=data['V1b'].value_counts()
print(value_counts_v1b)
```

- V1의 데이터 개수 (5287+5051+269) = 10607

- 공백값 = 2522
- V1b 의 공백값 = 10607
- => 중복되는 값이 없기에 numpy where을 사용하여 결합해도 데이터를 손상시키지 않을 것 
- V1 = '난민에 관한 우려' 로 제목 변경

```python
data['난민에 관한 우려'] = data['V1'].fillna(data['V1b'])
data.drop('V1', axis =1, inplace=True)
data.drop('V1b', axis =1, inplace=True)
```

```python
#v3의 값에 공백을 넣어둬서, 값으로 처리가 되어서 오류 발생했었음.. 전처리 확실히 하기
data['V3'] = data['V3'].replace(r'^\s*$', pd.NA, regex=True)
data['V3b'] = data['V3b'].replace(r'^\s*$', pd.NA, regex=True)
data['V3c'] = data['V3c'].replace(r'^\s*$', pd.NA, regex=True)
```
