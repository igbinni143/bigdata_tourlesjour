```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Selenium 드라이버 설정
driver = webdriver.Chrome()
driver.maximize_window()

# 데이터 저장용 리스트
store_data = []

# 기본 URL 설정
base_url = 'https://www.tlj.co.kr:7008/store/search.asp?page='

# 페이지 번호 설정
page = 1

while True:
    # 현재 페이지 URL로 이동
    url = base_url + str(page)
    driver.get(url)
    time.sleep(2)  # 페이지 로딩 대기
    
    # 매장 리스트 가져오기
    rows = driver.find_elements(By.CSS_SELECTOR, '#storeList tr')
    if not rows:  # 매장 리스트가 없으면 종료
        print("마지막 페이지입니다.")
        break

    for row in rows:
        try:
            # 매장명, 지역, 도로명 주소 추출
            store_name = row.find_element(By.CSS_SELECTOR, 'a').text
            area = row.find_element(By.CSS_SELECTOR, 'span[id^="listArea_"]').text
            road_address = row.find_element(By.CSS_SELECTOR, 'span[id^="listAddrD_"]').text
            
            # 데이터 추가
            store_data.append({
                '매장': store_name,
                '지역': area,
                '주소': road_address
            })
            print(store_data[-1])
        except Exception as e:
            print(f"Error while processing row: {e}")
    
    # 다음 페이지로 이동
    page += 1

# 데이터프레임 생성
df = pd.DataFrame(store_data)

# 데이터 확인
print(df)

# CSV 저장
df.to_csv('뚜레쥬르_매장정보.csv', index=False, encoding='utf-8-sig')

# 드라이버 종료
driver.quit()
```
